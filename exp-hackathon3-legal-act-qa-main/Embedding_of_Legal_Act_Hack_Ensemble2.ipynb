```python
from tqdm import tqdm
```

```python
tokenizer.model_max_length = 512
```

```python
model.config.output_hidden_states = True

batch_size = 1

text_list = []
word_embedding_list = []

for index, text in enumerate(tqdm(train["concat"])):
  text_list.append(text)
  index += 1
  if len(text_list)>=batch_size:
    with torch.no_grad():
      encoding = tokenizer(
        text_list,
        return_tensors='pt',
        padding = True,
        truncation=True,
        max_length = 508
      ).to(device)
      outputs = model(**encoding)
      word_embedding = outputs.hidden_states[-1][:, 0, :].cpu()
      word_embedding = torch.squeeze(word_embedding)
      word_embedding_list.append(word_embedding)
    text_list = []
```
